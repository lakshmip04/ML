```python
import numpy as np
import pandas as pd

# Define functions
def dimensionality(A):
    return A.shape[1]

def vector_space(A):
    return A.shape[0]

def rank(A):
    return np.linalg.matrix_rank(A)

def calculate_model_vector(A, C):
    A_pseudo_inv = np.linalg.pinv(A)
    X = A_pseudo_inv @ C
    return X



# Extract matrices A and C
A = data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].to_numpy()
C = data['Payment (Rs)'].to_numpy().reshape(-1, 1)

# Calculate and print results
dim = dimensionality(A)
print(f"Dimensionality of the vector space: {dim}")

num_vectors = vector_space(A)
print(f"Number of vectors in this vector space: {num_vectors}")

rank_A = rank(A)
print(f"Rank of Matrix A: {rank_A}")

# Calculate and print the model vector X
model_vector_X = calculate_model_vector(A, C)
print(f"Model vector X for predicting the cost of the products: {model_vector_X}")

# Display the cost of each product separately
product_names = ['Candies', 'Mangoes', 'Milk Packets']
for name, cost in zip(product_names, model_vector_X):
    print(f"Cost per unit of {name}: {cost[0]}")
```

```python
"""Mark all customers (in “Purchase Data” table) with payments above Rs. 200 as RICH and others
as POOR. Develop a classifier model to categorize customers into RICH or POOR class based on
purchase behavior."""
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score


data['Customer Class'] = np.where(data['Payment (Rs)'] > 200, 'RICH', 'POOR')
print(data.head())

X = data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']]
y = data['Customer Class']


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)


classifier = LogisticRegression()
classifier.fit(X_train, y_train)


y_pred = classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)


```

```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Colab Notebooks/LabSessionData.xlsx'
data = pd.read_excel(file_path, sheet_name='IRCTC Stock Price')

def calculate_mean_and_variance(data):
    price_data = data['Price'].dropna()
    mean_price = statistics.mean(price_data)
    variance_price = statistics.variance(price_data)
    return mean_price, variance_price

def calculate_wednesday_mean(data):
    data['Date'] = pd.to_datetime(data['Date'])
    data['DayOfWeek'] = data['Date'].dt.day_name()
    wednesday_data = data[data['DayOfWeek'] == 'Wednesday']
    wednesday_prices = wednesday_data['Price'].dropna()
    mean_wednesday = statistics.mean(wednesday_prices)
    return mean_wednesday

def calculate_april_mean(data):
    april_data = data[data['Date'].dt.month == 4]
    april_prices = april_data['Price'].dropna()
    mean_april = statistics.mean(april_prices)
    return mean_april

def calculate_probability_of_loss(data):
    chg_data = data['Chg%'].dropna()
    num_losses = sum(1 for change in chg_data if change < 0)
    prob_loss = num_losses / len(chg_data)
    return prob_loss

def calculate_probability_of_profit_on_wednesday(data):
    wednesday_data = data[data['DayOfWeek'] == 'Wednesday']
    num_wednesday_profits = sum(1 for change in wednesday_data['Chg%'].dropna() if change > 0)
    prob_profit_wednesday = num_wednesday_profits / len(wednesday_data)
    return prob_profit_wednesday

def calculate_conditional_probability_of_profit_given_wednesday(data):
    chg_data = data['Chg%'].dropna()
    num_total_profits = sum(1 for change in chg_data if change > 0)
    prob_profit = num_total_profits / len(chg_data)

    wednesday_data = data[data['DayOfWeek'] == 'Wednesday']
    num_wednesday_profits = sum(1 for change in wednesday_data['Chg%'].dropna() if change > 0)
    prob_profit_wednesday = num_wednesday_profits / len(wednesday_data)

    conditional_prob_profit_given_wednesday = (prob_profit_wednesday * prob_profit) / prob_profit
    return conditional_prob_profit_given_wednesday

def plot_chg_vs_day_of_week(data):
    plt.figure(figsize=(10, 6))
    days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
    data['DayOfWeek'] = pd.Categorical(data['DayOfWeek'], categories=days, ordered=True)
    plt.scatter(data['DayOfWeek'], data['Chg%'], alpha=0.5)
    plt.title('Chg% vs Day of the Week')
    plt.xlabel('Day of the Week')
    plt.ylabel('Chg%')
    plt.grid(True)
    plt.show()

mean_price, variance_price = calculate_mean_and_variance(data)
print(f"Mean of Price data: {mean_price}")
print(f"Variance of Price data: {variance_price}")

mean_wednesday = calculate_wednesday_mean(data)
print(f"Mean of Price data for Wednesdays: {mean_wednesday}")
print(f"Comparison with population mean: {mean_wednesday - mean_price}")

mean_april = calculate_april_mean(data)
print(f"Mean of Price data for April: {mean_april}")
print(f"Comparison with population mean: {mean_april - mean_price}")

prob_loss = calculate_probability_of_loss(data)
print(f"Probability of making a loss: {prob_loss}")

prob_profit_wednesday = calculate_probability_of_profit_on_wednesday(data)
print(f"Probability of making a profit on Wednesday: {prob_profit_wednesday}")

conditional_prob_profit_given_wednesday = calculate_conditional_probability_of_profit_given_wednesday(data)
print(f"Conditional probability of making a profit given it's Wednesday: {conditional_prob_profit_given_wednesday}")

plot_chg_vs_day_of_week(data)
```
